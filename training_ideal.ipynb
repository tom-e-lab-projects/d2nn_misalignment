Physical system parameters

```python
config = ss.OpticalConfig(
    532e-9,
    5e-6,
    256,
    256,
    4e-3
)

distance = 0.2 

# assume all distances the same
def distances(n):
    return distance*torch.ones(n+1).unsqueeze(0)
```

General training parameters

```python
train_size = 4096
val_size = 512
batch_size = 64
train_loader, val_loader = dg.data_set_generation(config.height, config.width, train_size, val_size, batch_size)
criterion = cc.npcc_loss
target_pos = 50
half_target_size = 10
target_method = tc.targets(config.height, config.width, target_pos, target_pos, half_target_size)
log_freq = 32
epoch_num = 50
```

Running single layer system

```python
d_1 = distances(1)
d2nn1 = pms.IdealSystemTraining(config, d_1 False)

optimizer = torch.optim.Adam(d2nn1.parameters(), lr=1e-2)

path_1 = 'ideal_training/d2nn1'

d2nn1, train_losses, train_accies, val_losses, val_accies, I_val, labels_val =\
        tv.train(d2nn1,criterion, optimizer,
                train_loader,val_loader, 
                target_method,
                path_1, log_freq, epoch_num)  
```

2 layer

```python
d_2 = distances(2)
d2nn2 = pms.IdealSystemTraining(config, d_2, False)
optimizer = torch.optim.Adam(d2nn1.parameters(), lr=1e-2)
path_2 = 'ideal_training/d2nn2'

d2nn2, train_losses, train_accies, val_losses, val_accies, I_val, labels_val =\
        tv.train(d2nn1,criterion, optimizer,
                train_loader,val_loader, 
                target_method,
                path_2, log_freq, epoch_num)  
```
